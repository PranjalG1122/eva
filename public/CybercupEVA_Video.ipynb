```python
def gen(text):
  hash = hashlib.md5(text.encode()).hexdigest()
  !tts --text "{text}" --model_name "{model}" --out_path "/content/audio_data/{hash}.wav"
  !python /content/SadTalker/inference.py \
                    --driven_audio "/content/audio_data/{hash}.wav" \
                    --source_image "{image}" \
                    --result_dir "{result_dir}" \
                    --result_name "{hash}" \
                    --still \
                    --preprocess full
  !ffmpeg -i "/content/static/results/{hash}.mp4" -c:v libvpx-vp9 -crf 15 -b:v 0 -b:a 128k -c:a libopus "/content/static/results/{hash}.webm"
  return hash
```

# Flask endpoint

```python
# break
app = Flask(__name__)
run_with_ngrok(app=app, auth_token='')

@app.route('/', methods=['POST', 'GET'])
def hello_world():
    print(request)
    print(request.form)
    print(request.method)

    hash = gen(request.form["text"])

    return f"/static/results/{hash}.webm"

os.environ["FLASK_DEBUG"] = "False"
app.run()
```
